```python
class TreeNode {
 int value;
 Node* left;
 Node* right;
};

bool twoElementsEqualTarget(TreeNode *root, int target) {
    // inorder traversal
    vector<int> path;
    traverse(root, path);
    int lo = 0, hi = path.size() - 1;
    while (lo < hi) {
        int sum = path[lo] + path[hi];
        if (sum < target) ++lo;
        else if (sum > target) --hi;
        else return true;
    }
    return false;
}

void traverse(TreeNode *root, vector<int> &path) {
    if (root == nullptr) return;
    if (root->left) traverse(root->left, path);
    path.push_back(root->value);
    if (root->right) traverse(root->right, path);
}


class Solution {
    unordered_map<int, Node *> value2Node;
    public Node *CloneGraph(Node *graph) {
        if (graph == nullptr) return nullptr;
        if (value2Node.count(graph->val)) return value2Node[graph->val];
        Node *newGraph = new Node(graph->val);
        value2Node[graph->val] = newGraph;
        for (Node *neighbor: graph->neighbors) {
            (newGraph->neighbors).push_back(CloneGraph(neighbor));
        }
        return newGraph;
    }
};
```
